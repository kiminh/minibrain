```python
def get_vals(f):
    cols = [ 'UPOS', 'XPOS', 'FEATS', 'HEAD', 'DEPREL', 'DEPS', 'MISC']
    dictvals = {}

    for n in cols:
        if not n in dictvals:
            try:
                dictvals[n] = set(np.unique(f[n].values))
            except:
                pass
        else:
            try:
                dictvals[n] = dictvals[n].union(set(np.unique(f[n].values) ) )
            except:
                pass
    return dictvals
```

```python
import conllu
import ntpath
# from https://stackoverflow.com/questions/8384737/extract-file-name-from-path-no-matter-what-the-os-path-format
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)
```

```python
#fname = "/home/leo/projects/Datasets/text/UniversalDependencies/ud-treebanks-v2.4/UD_French-PUD/fr_pud-ud-test.conllu"
def get_conllu(fname):
    f = open(fname,"r")
    cnl = conllu.parse(f.read())
    return path_leaf(fname), cnl
```

```python
# fields according to conllu parser lib
#fields = ['id', 'form', 'lemma', 'upostag', 'xpostag', 'feats', 'head', 'deprel', 'deps', 'misc']
fields = ['upostag', 'xpostag', 'feats', 'deprel']
```

```python
def filter_fields(conllu_dict, fields):
    """
    filter only the wanted fields from the conllu input dictionary
    """
    accum = {}
    for t in conllu_dict:  # each item is a line in a TokenList object
        for d in t:  # each item is a word/symbol
            for f in fields:  # only get the fields I need
                if f in accum:  #  and accum[f] is not None:
                    print(f, accum[f], d[f], accum[f], d[f])
                    accum[f] = accum[f].append(d[f])
                else:
                    lst = []
                    lst.append(d[f])
                    accum[f] = lst
    return accum
```

```python
def get_conllu_tag_values(fname, fields):
    
    f = open(fname,"r")
    fid, parsed = get_conllu(fname)
    ret = (fid, filter_fields(parsed, fields))
    return ret
```

```python
def get_tags(parsed):
    upostags = []
    xpostags = []
    feattags = []
    depreltags = []

    for t in parsed:
        for d in t:
            upostags.append(d['upostag'])
            xpostags.append(d['xpostag'])
            fts = []
            if d['feats'] is not None:
                fts = list(d['feats'].items())
                feattags.extend(fts)
            depreltags.append(d['deprel'])
            #feats are OrderedDicts
#             lfeattags = [(f.items) for f in feattags if f is not None]
    return set(upostags), set(xpostags), set(feattags), set(depreltags)
```

```python
def get_conllu_tag_values(fname):
    f = open(fname,"r")
    fid, parsed = get_conllu(fname)
    ret = get_tags(parsed)
    return fid, ret
```
